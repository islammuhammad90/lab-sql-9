```python
import pandas as pd
import numpy as np

import pymysql
from sqlalchemy import create_engine

from getpass import getpass  # To get the password without showing the input
```

```python
connection_string = 'mysql+pymysql://root:'+password+'@localhost/sakila'
engine = create_engine(connection_string)
```

```python
#1 How many distinct (different) actors' last names are there?

query1 = 'select count(last_name) as unique_last_names \
from actor \
group by last_name \
having count(last_name) = 1 '

data1 = pd.read_sql_query(query1, engine)
data1.head(100)
display(data1.shape)
```

```python
#2 Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.

query2 = 'select date(rental_date) as "rental_date" , date_format(rental_date, "%%W") AS weekday, \
case when date_format(rental_date, "%%W") IN ("Saturday","Sunday")  THEN "weekend" ELSE "weekday" END as "weekday_or_weekend" \
from rental '

# date_format(rental_date, "%%W") = "Saturday" THEN "weekend" ELSE "weekday" END as "weekday_or_weekend"
# date_format(rental_date, "%%W") AS weekday
data2 = pd.read_sql_query(query2, engine)
data2.head(60)
##display(data2.shape)
```

```python
#3 Get all films with ARMAGEDDON in the title.

query3 = 'select title as "title" \
from film \
where title like "%%Armageddon%%" '

data3 = pd.read_sql_query(query3, engine)
data3.head(20)
#display(data3.shape)
```

```python
#4 Get 10 the longest films.

query4 = 'select title as "title", length as duration \
from film \
order by 2 desc \
limit 10'

data4 = pd.read_sql_query(query4, engine)
data4.head(20)
#display(data4.shape)
```

```python
#5 How many films include Behind the Scenes content?

query5 = 'select "films_with_behind_scenes" as "type", count(title) as "films_w_behindscene" \
from film \
where special_features like "%%behind the scenes%%" '

data5 = pd.read_sql_query(query5, engine)
data5.head()
#display(data5.shape)
```

```python
#6 Which kind of movies (rating) have a mean duration of more than two hours?

query6 = 'select rating, count(film_id) as "films", round(avg(length),2) as "avg_duration" \
from film \
group by rating \
having avg(length) > 120'


data6 = pd.read_sql_query(query6, engine)
data6.head()
#display(data6.shape)
```

```python
#7 Rank films by length (filter out the rows that have nulls or 0s in length column).
#In your output, only select the columns title, length, and the rank.

query7 = 'select title, length, DENSE_Rank() OVER(order by length desc) AS "rank" \
from film \
where length IS NOT NULL and length > 0 '


data7 = pd.read_sql_query(query7, engine)
data7.head(30)
#display(data7.shape)
```
